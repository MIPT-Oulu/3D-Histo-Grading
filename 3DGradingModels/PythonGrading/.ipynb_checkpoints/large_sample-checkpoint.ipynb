```python
import torch
import torch.nn as nn
import torchvision
from model import UNet

def segmentation_pytorch(sample, modelpath, mu=108.58790588378906, sd=47.622276306152344):
    #Load model
    model = UNet(24,6,1,1)
    model.load_state_dict(torch.load(modelpath))
    model = model.cuda()

    #Sample dimensions
    dims = sample.shape
    bs = 10
    
    #Empty array for output
    output = np.zeros((dims[0],dims[1],512))
    #Inference over XZ projections
    for k in range(0,dims[0],bs):
        k = int(k/bs)
        #Minibatch
        batch = sample[k*bs:(k+1)*bs,:,:512]
        #Batch to tensor        
        T = torch.from_numpy(batch).float()        
        #Set dimensiosn
        try:
            T = T.view(bs,1,dims[1],512)
        except RuntimeError:
            break
        T = (T-mu)/sd
        #Inference

        #Sigmoid will scale the result to range [0,1]
        SM = nn.Sigmoid()
        with torch.no_grad():
            out = SM(model(T.cuda()))

        #Inference to numpy
        out = out.cpu().data.numpy()
        out = out.reshape(bs,dims[1],512)
        
        output[k*bs:(k+1)*bs,:,:] = out


    return output

mask = segmentation_pytorch(data_crop, r"Z:\Santeri\3DGradingModels\PythonGrading\snapshots\2018_11_16_00_05\fold_2_epoch_31.pth")
PrintOrthogonal(mask)
```
